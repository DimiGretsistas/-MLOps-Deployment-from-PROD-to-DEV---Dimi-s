```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# 1. Create data
# Relationship: y = 2x + noise
X = np.array([[i] for i in range(100)])
y = 2 * X.flatten() + np.random.normal(0, 5, 100)

# 2. Convert to DataFrame for easier handling
df = pd.DataFrame({'Feature': X.flatten(), 'Target': y})

# 3. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['Feature']], df['Target'], test_size=0.2, random_state=42)

# 4. Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# 5. Make a simple prediction
# We use a DataFrame with the same column name to avoid the UserWarning
new_data = pd.DataFrame({'Feature': [10]})
prediction = model.predict(new_data)

print(f"Model trained. Prediction for value 10: {prediction[0]:.2f}")
```
